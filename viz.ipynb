```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
with open("output/output.csv", "r") as f:
    result_csv = f.readlines()    
```

```python
result_csv
```

```python
i = 1
plt.figure(figsize=(30, 10))
for line in result_csv[:10]:
    name, pred_xmin, pred_ymin, pred_xmax, pred_ymax, xmin, ymin, xmax, ymax = line.split(",")
    ymax = ymax[:-1]
    img_path = os.path.join("one_revoc/img", name)
    
    blank = cv2.imread(img_path)
    pred = cv2.rectangle(blank, (int(pred_xmin), int(pred_ymax)), (int(pred_xmax), int(pred_ymin)), (255, 255, 255), 2)
    label = cv2.rectangle(blank, (int(xmin), int(ymax)), (int(xmax), int(ymin)), (255, 255, 255), 2)

    plt.subplot(10, 3, i)
    plt.axis("off")
    plt.figure(figsize=(3, 3))
    plt.imshow(blank)
    i+=1
    plt.subplot(10, 3, i)
    plt.axis("off")
    plt.figure(figsize=(3, 3))
    plt.imshow(pred)
    i+=1
    plt.subplot(10, 3, i)
    plt.axis("off")
    plt.figure(figsize=(3, 3))
    plt.imshow(label)
    i+=1

    
```
